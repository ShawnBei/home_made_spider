```python
import requests
url = "https://www.1905.com/vod/list/n_1_a_4/o3p1.html"
response = requests.get(url)
print(response.status_code)
print(response.content.decode('utf-8'))
# 保存网页到本地
file_obj = open('1905.html', 'w')  # 以写模式打开名叫 douban.html的文件
# 如果打开网页显示的是乱码那么就用下一行代码
# file_obj = open('douban.html', 'w', encoding="utf-8")  # 以写模式打开名叫 douban.html的文件，指定编码为utf-8
file_obj.write(response.content.decode('utf-8'))  # 把响应的html内容
file_obj.close()  # 关闭文件，结束写入
```

```python
from bs4 import BeautifulSoup
#读取文件内容到 html 变量里面
file_obj = open('1905.html','r') #以读的方式打开文件
html = file_obj.read()
file_obj.close()

soup = BeautifulSoup(html,'lxml')
# print(soup)
```

```python
import csv 
csv_file = open('movie_data.csv','w',encoding="UTF-8",newline='')
writer = csv.writer(csv_file)

writer.writerow(["Movie Name","Link","Score","Description"])

all_movies = soup.find('div',id="content")
for each_movie in all_movies.find_all('div',class_="grid-2x grid-3x-md grid-6x-sm"):
        #print(each_movie)
        all_a_tag = each_movie.find_all('a')
        movie_name = each_movie.find('h3').text
        movie_url = all_a_tag[0]['href']
        movie_score = each_movie.find('i').text
        movie_des = each_movie.find('p').text
        writer.writerow([movie_name,movie_url,movie_score,movie_des])

        
        
csv_file.close()
print("finished!")
```
